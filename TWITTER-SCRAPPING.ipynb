# Scraping JIOMART customer feed back tweets from Twitter

```python
# Importing sntwitter for scraping the data
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data to
jiomart_tweet_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('jiomart tweet support since:2019-01-01 until:2021-12-31').get_items()):
    if i>5000:
        break
    jiomart_tweet_list.append([tweet.date, tweet.id,tweet.user.username,tweet.content,tweet.likeCount,tweet.replyCount])
    
# Creating a dataframe from the tweets list above
jiomart_df = pd.DataFrame(jiomart_tweet_list, columns=['TimeStamp','Tweet Id','Name','Tweet','LIke','ReplyCount'])
```

```python
#shape of the dataframe
jiomart_df.shape
```

```python
# Removing Duplicate values
jiomart_df.drop_duplicates(inplace=True)
```

```python
jiomart_df.shape
```

```python
jiomart_df['COMPANY NAME'] = 'JIOMART'
```

```python
jiomart_df
```
